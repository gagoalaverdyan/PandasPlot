```python
import pandas as pd
import numpy as np

my_table = pd.read_csv("dog_data.csv")
my_table
```

```python
# Conditional selection
# < > == <= >=
my_table == "BROWN"
```

```python
# Return dataframe with data
my_table[my_table == "BROWN"]
```

```python
# Return only the needed column
my_table[my_table == "BROWN"]["Color"]
```

```python
# Return only the needed columns
my_table[my_table == "BROWN"][["Color", "Breed"]]
```

```python
# Multiple conditional AND
my_table[(my_table["Color"] == "GREY") & (my_table["Breed"] == "MIXED")]
```

```python
# Multiple conditional OR
my_table[(my_table["Color"] == "GREY") | (my_table["Breed"] == "MIXED")]
```

```python
# Count of the multiple conditional selection
len(my_table[(my_table["Color"] == "GREY") | (my_table["Breed"] == "MIXED")])
```

```python
# Return specific column of multiple conditional selection
my_table[(my_table["Color"] == "GREY") | (my_table["Breed"] == "MIXED")]["DogName"]
```
