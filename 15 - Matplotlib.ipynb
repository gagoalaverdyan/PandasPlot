```python
import pandas as pd
import numpy as np
from numpy.random import randn
%matplotlib inline

my_df = pd.DataFrame(randn(100, 4), columns=["Man", "Van", "Can", "Han"])
```

```python
# Histogram
my_df["Han"].hist()
```

```python
# Historgam with smaller bins
my_df["Han"].hist(bins=40, grid=False, legend=True)
```

```python
# No grid bars
my_df["Can"].plot(kind="hist", bins=20)
```
