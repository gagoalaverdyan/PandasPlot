```python
import pandas as pd
import numpy as np
from numpy.random import randn
%matplotlib inline

my_df = pd.DataFrame(randn(400, 4), columns=["Man", "Van", "Can", "Han"])
```

```python
# Area plot
my_df[["Man", "Han"]].abs().plot(
    kind="area", alpha=0.6, title="My first plot!", legend=False
)
```

```python
# Bar plot
my_df.abs().plot(kind="bar", stacked=True, alpha=0.6, grid=False, title="My aws0me barchart :)")
```

```python
# Line plot
my_df.abs().plot(kind="line", lw=2, title="Heheheheheeeeeee")
```

```python
# Scatter plot
my_df.plot(kind="scatter", x="Man", y="Van", c="Han", colormap="magma")
```

```python
# Box plot
my_df.plot(kind="box", figsize=(10, 5), colormap="cool")
```

```python
# Hex Plot
my_df.plot(kind="hexbin", x="Man", y="Han", gridsize=20, cmap="Blues", title="Bluecomb")
```

```python
# Density Plot
my_df.plot(kind="density", lw=2, title="KDE Plot")
```
